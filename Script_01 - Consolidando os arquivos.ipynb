```python
# Importar as pacores/biblitoecas/Libs

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

# Acessar compenentes do SISTEMA OPERACIONAL
import os
from pathlib import Path
```

```python
# Local direotiro
Local_Diretorio = 'C:/Users/datav/Desktop/Projeto_Dashboard_Evento/infrações_2022/ajustados_2022'
Diretorio_Path = Path(Local_Diretorio)

type(Diretorio_Path)
```

```python
# Lista os arquivos
Arquivos = Diretorio_Path.iterdir()

# Função para calcular o tamanho do arquivo
def Bytes_Mega(bytes):
    Megabytes = bytes / ( 1024 * 1024)
    return Megabytes

# Total MBs
Total_MB = 0

# Loop
for Loop in Arquivos:
    
    # Buscando o tamanho do arquivo
    Tamanho = os.path.getsize( Loop )
    
    # Convervento em Megas
    Conversao = Bytes_Mega( Tamanho )
    
    # Extraindo o nome do arquivo
    Nome_Arquivo = str( Loop ).split('\\')[-1]
    
    print(f' {Nome_Arquivo} : { round(Conversao, 2) } MB')
    print('-' * 30 )
    
    Total_MB = Total_MB + Tamanho
    
print( f'Temos ao todo { round( Bytes_Mega(Total_MB), 2) } MBs de dados' )
```

```python
# Amostra dos dados
pd.read_csv( Loop, error_bad_lines=False, sep=';', encoding='latin-1'  ).head()
```

```python
# Lista os arquivos
Arquivos = Diretorio_Path.iterdir()

# Loop para consolidar os arquivos
for Contador, Arq in enumerate(Arquivos):
    
    print(Contador, Arq )
    
    # Se for a 1º interação, cria a base consolidada
    if Contador == 0:
        # Ler o arquivo do loop
        Base_Dados = pd.read_csv( Arq, error_bad_lines=False, sep=';', encoding='latin-1' )
        # Salvar na var 
        Base_Consolidada = Base_Dados
        
    else:
        # Ler o arquivo do loop
        Base_Dados = pd.read_csv( Arq, error_bad_lines=False, sep=';', encoding='latin-1' )
        # Empilhar os dados
        Base_Consolidada = pd.concat( [Base_Consolidada, Base_Dados] )
    
# Dimensão
Base_Consolidada.shape
```

```python
Base_Consolidada.tail()
```

```python
# Exportação, CSV
Base_Consolidada.to_csv('Base_Consolidada.csv', index=False)
```
